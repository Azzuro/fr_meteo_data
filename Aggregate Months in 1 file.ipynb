# Aggregate in a full year

```
year = "2019"
df_year = pd.DataFrame()
for month in range(12):
  smonth = year + "-" + str(month+1).rjust(2, '0')
  print ("Month: ", smonth)
  try:
    df_month = pd.read_csv(root + "nr_" + smonth + ".csv")
    # concat with the year df
    df_year = pd.concat([df_year, df_month])
  except:
    break

df_year.to_csv(root + year + ".csv")
```
